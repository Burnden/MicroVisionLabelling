```python
import imgaug as ia
from imgaug import augmenters as iaa
import numpy as np
from matplotlib import pyplot as plt
import skimage
import skimage.io
import csv
import os.path
%matplotlib inline
```

```python
img_path = r"C:\Users\wuyua\Desktop\MicroVisionLabelling\Dataset\Batch 4\1523393527797.png"
img = skimage.io.imread(img_path)
plt.gray()
plt.imshow(img)
```

```python
# load annotation

annotation = [];
target_file_path, target_file_name = os.path.split(img_path)
mask_file = os.path.join(target_file_path, "via_region_data.csv")

with open(mask_file, "r") as csvfile:
    reader = csv.reader(csvfile)
    next(reader)
    annotation = [x[5] for x in reader if x[0] == target_file_name]
```

```python
# create mask
mask = np.zeros(img.shape)
for ann in annotation:
    skimage.draw.circle(ann)
```
